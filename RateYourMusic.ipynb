```python
def getRYMData(files):
    rymData = []
    for i,ifile in enumerate(files):
        print(ifile)
        bsdata = getHTML(ifile)

        table = bsdata.find("table", {"id": "user_list"})
        if table is None:
            print("Could not find user_list table!")
            continue

        tableData = []
        trs = table.findAll("tr")
        for itr,tr in enumerate(trs):
            tds = tr.findAll("td")
            print(itr,'\t',len(tds))
            if len(tds) == 4:
                posData    = tds[0]
                imgData    = tds[1]
                musicData  = tds[2]
            elif len(tds) == 3:
                posData    = tds[0]
                imgData    = tds[1]
                musicData  = tds[2]
            else:
                raise ValueError("Not sure what to do with {0}".format(tr))
                
            print(musicData)
            continue
                

            ###################################
            ## Title
            ###################################
            try:
                ref  = imgData.find("a")
                titleURL  = ref.attrs['href']
            except:
                titleURL = None

            try:
                img  = imgData.find("img")
                titleName = img.attrs['title']
            except:
                titleName = None
            titleData = {"Name": titleName, "URL": titleURL}


            ###################################
            ## Artist
            ###################################
            artistRef = musicData.find("a", {"class": "list_artist"})
            if artistRef is not None:
                artistName  = artistRef.text
                artistURL   = artistRef.attrs['href']
                artistTitle = artistRef.attrs['title']
            else:
                artistName  = None
                artistURL   = None
                artistTitle = None
            artistData = {"Name": artistName, "URL": artistURL, "Title": artistTitle}


            ###################################
            ## Album
            ###################################
            albumRef  = musicData.find("a", {"class": "list_album"})
            if albumRef is not None:
                albumName  = albumRef.text
                albumURL   = albumRef.attrs['href']
                albumTitle = albumRef.attrs['title']
            else:
                albumName  = None
                albumURL   = None
                albumTitle = None
            albumData = {"Name": albumName, "URL": albumURL, "Title": albumTitle}
            tableData.append({"Artist": artistData, "Album": albumData, "Title": titleData})
            print(tableData)

        rymData.append(tableData)
    return rymData
```
