```python
def downloadURL(url):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,} 

    print("Now Downloading {0}".format(url))

    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
        
    return data, response.getcode()
```

```python
from pandas import date_range
#2016-12-23--2016-12-30
#2017-01-06--2017-01-13
from datetime import datetime
start = '12/23/2016'
end   = datetime.now().strftime("%Y-%m-%d")
datevals = [x.strftime('%Y-%m-%d') for x in list(date_range(start=start, end=end, freq='W-FRI'))]
```

```python
from time import sleep
from fsUtils import isFile

for country in countries:
    for i,dateval in enumerate(datevals):
        if i == len(datevals)-1:
            break
        start = datevals[i+1]
        end   = datevals[i]
        url      = "https://spotifycharts.com/regional/{0}/weekly/{1}--{2}/download".format(country, start, end)
        savename = "spotify/regional-{0}-weekly-{1}--{2}.p".format(country, start, end)
        if isFile(savename):
            continue

        try:
            data, code = downloadURL(url)
        except:
            break
        saveFile(idata=data, ifile=savename)
        sleep(3.5)
```

```python
years = range(1900, 2020+1)
for year in years:
    url="https://www.musicvf.com/top_songs_of_{0}".format(year)
    savename = "musicvf/{0}.p".format(year)
    print(savename)
    bsdata = getHTML(savename)
    
    _ = mkDir("musicvf/{0}".format(year))
    saveDir = setDir("musicvf", str(year))

    for div in bsdata.findAll("div", {"id": "navcontainer"}):
        uls = div.findAll("ul", {'id': 'navlist'})
        for ul in uls:
            lis = ul.findAll("li")
            for li in lis:
                href = li.find('a').attrs['href']
                name = li.find('a').text

                if name == "Singles / Songs Top 100":
                    continue
                elif name == "Artists Top 50":
                    savename = "musicvf/{0}/{1}.p".format(year, "Artists")
                elif name == "R&B Songs Top 100":
                    savename = "musicvf/{0}/{1}.p".format(year, "RnB")   
                elif name == "Country Songs Top 100":
                    savename = "musicvf/{0}/{1}.p".format(year, "Country")
                elif name == "AC Songs Top 100":
                    savename = "musicvf/{0}/{1}.p".format(year, "Adult")                    
                elif name == "Dance Songs Top 100":
                    savename = "musicvf/{0}/{1}.p".format(year, "Dance")
                elif name == "Rock Songs Top 100":
                    savename = "musicvf/{0}/{1}.p".format(year, "Rock")
                else:
                    raise ValueError("Unknown name: {0}".format(name))

                url="https://www.musicvf.com/{0}".format(href)
                data, code = downloadURL(url)
                saveFile(idata=data, ifile=savename)
                sleep(3)
```

```python
savename = "musicvf.p"
bsdata = getHTML(savename)

divLetters = ["bloc-{0}".format(x) for x in ["A", "B", "C", "D", "E", "F", "H"]]
headers    = [bsdata.find("div", {"class": "{0}".format(x)}) for x in divLetters]
headers    = [x.text.replace("\xa0", "").strip() for x in headers]
data = {}
retval = {}
for i,divLetter in enumerate(divLetters):
    if headers[i] == "USpeak":
        continue
    entries = bsdata.findAll("div", {"class": "{0}".format(divLetter)})[1:]
    data[headers[i]] = entries

#data["Year"]  = [x.find('a').text for x in data["Date"]]
retval = []
for j,val in enumerate(data["TitleAlbum"]):
    #print(j,'\t',len(val.findAll("a")))
    taHeaders = ["Song", "Artist", "Album", "Image"]
    taRefs    = val.findAll('a')
    taData    = dict(zip(taHeaders, taRefs))
    #print(taData)

    try:
        titleName = taData["Song"].attrs['title']
        titleName = titleName.replace("[article][video]", "").strip()
        titleURL  = taData["Song"].attrs['href']
    except:
        titleName = None
        titleURL  = None
    titleData = {"Name": titleName, "URL": titleURL}
    #print('\t',titleData)
    
    try:
        artistName = taData["Artist"].text
        artistURL  = taData["Artist"].attrs['href']
    except:
        artistName = None
        artistURL  = None
    artistData = {"Name": artistName, "URL": artistURL}
    #print('\t',artistData)
    
    try:
        albumName  = taData["Album"].attrs['title']
        albumURL   = taData["Album"].attrs['href']
    except:
        albumName  = None
        albumURL   = None
    albumData  = {"Name": albumName, "URL": albumURL}
    #print('\t',albumData)
    
    songData = {"Title": titleData, "Artist": artistData, "Album": albumData} #, "Year": data['Year'][j]}
    print(j,'\t',songData)
    retval.append(songData)
```
