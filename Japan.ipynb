```python
def downloadURL(url):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,} 

    print("Now Downloading {0}".format(url))

    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
        
    return data, response.getcode()
```

```python
def getRankerData(bsdata):
    retval = {}
    for script in bsdata.findAll("script"): #, {"type": "application/ld+json"})
        if len(script.contents) > 0:
            try:
                jdata = json.loads(script.contents[0])
            except:
                continue

            if script.attrs.get('id') is not None:
                continue

            #print(script.attrs)
            title=jdata[1]['name']
            retval["Title"] = title
            artists = []
            for item in jdata[1]['itemListElement']:
                artist = item.get('name')
                if artist is None:
                    continue
                artists.append(artist)
            retval["Artists"] = artists

    return retval
```
