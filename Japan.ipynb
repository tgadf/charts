```python
def downloadURL(url):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,} 

    print("Now Downloading {0}".format(url))

    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
        
    return data, response.getcode()
```

```python
from searchUtils import findExt
from webUtils import getHTML
import json
from fsUtils import mkDir, setDir, isFile, setFile
from fileUtils import getBaseFilename


files = findExt("deezer/", ".p")
for ifile in files:
    bsdata = getHTML(ifile)
    
    jdata = None
    for script in bsdata.findAll("script"):
        if len(script.contents) == 0:
            continue

        if script.contents[0].startswith("window.__DZR_APP_STATE__ = "):        
            try:        
                jdata = json.loads(script.contents[0].replace("window.__DZR_APP_STATE__ = ", ""))
            except:
                continue

        if jdata is not None:
            break

    if jdata is not None:
        saveDir = setDir("deezer", getBaseFilename(ifile))
        mkDir(saveDir)
    
    
    for item in jdata['sections']:
        playlists = item['items']
        for playlist in playlists:
            target = playlist['target']
            title  = playlist['title']
            print(target,'\t',title)

            url = "https://www.deezer.com/en{0}".format(target)
            savename = setFile(saveDir, "{0}.p".format(title.replace(" ", "_")))
            if isFile(savename):
                continue

            data, code = downloadURL(url)
            saveFile(idata=data, ifile=savename)
            sleep(3)
```
